```python
from ultralytics import YOLO

model = YOLO(r"C:\Users\farra\Downloads\hfdyolov11\runs\weights\best.pt") # load a custom model
```

```python
img = r"C:\Users\farra\Downloads\images\image_2489.jpg"
```

```python
# Loop through each result and plot the detections
for result in results:
    img_with_boxes = result.plot()  # Annotate the image with bounding boxes
    plt.imshow(img_with_boxes)  # Display the image with annotations
    plt.show()  # Show the image
```

```python
import cv2
from ultralytics import YOLO

model = YOLO(r"C:\Users\farra\Downloads\hfdyolov11\runs\weights\best.pt")
cap = cv2.VideoCapture(0)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
while cap.isOpened():
    success, frame = cap.read()
    if success:
        results = model.track(frame, persist=True)
        annotated_frame = results[0].plot()
        cv2.imshow("YOLO11 Tracking", annotated_frame)
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```
